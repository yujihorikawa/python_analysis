```python
import matplotlib.pyplot as plt
import numpy as np

# 
# 
y0 = [1, 2, -5, 2]

# sin
x_sin = np.linspace(0, 10, 100)
y_sin = 2 + 2 * np.sin(2 * x_sin)

# 
x_hist = np.random.normal(15, 5, 2000)

# 
np.random.seed(3)
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))

# 
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 
axes[0, 0].plot(y0, marker='o')
axes[0, 0].set_title("Line Graph")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Value")

# sin
axes[0, 1].plot(x_sin, y_sin, linewidth=2.0)
axes[0, 1].set_title("Sine Function")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("y")

# 
axes[1, 0].hist(x_hist, bins=30, alpha=0.7)
axes[1, 0].set_title("Histogram")
axes[1, 0].set_xlabel("Value")
axes[1, 0].set_ylabel("Frequency")

# 
axes[1, 1].scatter(x_scatter, y_scatter, alpha=0.6)
axes[1, 1].set_title("Scatter Plot")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("y")

# 
plt.tight_layout()
plt.show()
```
